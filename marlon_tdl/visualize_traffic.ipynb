```python
import pandas as pd
import numpy as np
```

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse("nets/bo/joined_busses.rou.xml")
root = tree.getroot()
```

```python
vehicles = root.findall("vehicle")
```

```python
joined_busses = []
for vehicle in vehicles:
    joined_busses.append(vehicle.get('depart'))
```

```python
tree = ET.parse("nets/bo/joined.rou.xml")
root = tree.getroot()
```

```python
vehicles = root.findall("vehicle")
```

```python
len(vehicles)
```

```python
joined = []
for vehicle in vehicles:
    joined.append(vehicle.get('depart'))
```

```python
import matplotlib.pyplot as plt
```

```python
all_vehicles = []
all_vehicles.extend(joined)
all_vehicles.extend(joined_busses)
all_vehicles = sorted(all_vehicles)
```

```python
len(all_vehicles)
```

```python
plt.plot([i for i in range(len(all_vehicles))],all_vehicles)
plt.show();
```

```python
plt.hist(all_vehicles)
plt.show();
```
